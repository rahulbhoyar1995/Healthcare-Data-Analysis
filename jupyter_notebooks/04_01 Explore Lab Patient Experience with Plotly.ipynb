```python
import pandas as pd
import plotly.express as px
```

```python
df_lab = pd.read_csv("/content/demo_patient_experience_data.csv")
df_lab
```

```python
df_lab = df_lab.drop("Unnamed: 0", axis=1)
df_lab
```

```python
df_lab.info()
```

```python
px.bar(df_lab,x="Test Type", title="Test Type Distribution", color ="Test Type")
```

```python
fig = px.box(df_lab,x="Staff Rating", y="Speed Rating", title="Staff vs Speed Ratings")
fig.show()
```

```python
px.violin(df_lab,x="Wait Time (minutes)", title = "Wait Time distribution")
```

```python
px.violin(df_lab,x="Wait Time (minutes)", title = "Wait Time distribution", box=True)
```

```python
px.violin(df_lab,x="Wait Time (minutes)", title = "Wait Time distribution", box=True, points="all")
```
