```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_visit['Gender'].value_counts()
```

```python
visit_gender = df_visit['Gender'].value_counts()
```

```python
visit_gender.plot()
```

```python
visit_gender.plot(kind='bar')
```

```python
visit_gender.plot(kind='bar')
plt.title("Visit Gender Distribution")
```

```python
visit_gender.plot(kind='bar',color=['blue','pink'])
plt.title("Visit Gender Distribution")
```

```python
visit_gender.plot(kind='bar',color=['blue','pink'])
plt.title("Visit Gender Distribution")
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
visit_gender.plot(kind='bar',color=['blue','pink'])
plt.title("Visit Gender Distribution")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
